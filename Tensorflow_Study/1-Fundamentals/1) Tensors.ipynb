### Shuffle the order of the elements in a tensor

```python
# Shuffle a tensor(valuable for when you want to shuffle your data)
```

```python
not_shuffled = tf.constant([[7, 10],
                            [2, 3],
                            [12, 8]])
```

```python
tf.random.shuffle(not_shuffled)  # shuffle dimension 0 "prevent row"
```

```python
tf.random.shuffle(not_shuffled, seed = tf.random.set_seed(42))
```

```python
tf.random.set_seed(42) # global random seed
print(tf.random.shuffle(not_shuffled, seed = 21)) # operational random seed
print(tf.random.shuffle(not_shuffled, seed = 82))
```

### @tf.function

```python
def function(x, y):
    return x ** 2 + y

x = tf.constant(np.arange(0, 10))
y = tf.constant(np.arange(10, 20))

function(x, y)
```

```python
@tf.function
def tf_function(x ,y):
    return x ** 2 + y

tf_function(x, y)
```
